```python
class Solution(object):
    def isValid(self, s):
        mat_pair = {')':'(', '}':'{', ']':'['}
        stack = []
        for char in s:
            if char in '({[':
                stack.append(char)
            elif char in ')}]':
                if not stack or stack[-1] != mat_pair[char]:
                    return False
                stack.pop()
        return not stack
        """
        :type s: str
        :rtype: bool
        """
        
```
